###### 2 - Um computador em que as células e as palavras são de 12 bits

### Inicialmente temos as importações que foram utilizadas:

```python
import math, random
```

### Logo mais, temos as informações para a memória Cache:

```python
cMP = 16 # Capacidade da MP 
qtdLinha = 4 # Quantidade de linha
linhaConjunto = 2 # Quantidade de linha do conjunto
largura = 2 # largura do bloco / linha

qtdBloco = cMP / largura # Quantidade de bloco <---- 8
qtdConjunto = qtdLinha / linhaConjunto # Quantidade de conjunto <---- 2
qtdBlocoConjunto = qtdBloco / qtdConjunto # Quantidade de bloco por conjunto <---- 4

larguraCampoTag =int(math.log2(qtdBlocoConjunto)) # Largura do campo tag <---- 2
larguraCampoConjunto = int (math.log2(qtdConjunto)) # Largura do campo conjunto <---- 1
larguraCampoByte = int (math.log2(largura)) # Largura do campo Byte(célula) <---- 1
```

### Após isso, declaramos as Strings que serão utilizadas no decorrer do código quando for feito a quebra de Endereço em Tag, Conjunto e Byte:

```python
Ltag = "" # String que recebe a parte do endereço do campo tag 
Lconjunto = ""  # String que recebe a parte do endereço do campo conjunto
Lbyte = ""  # String que recebe a parte do endereço do campo byte 
```

### A seguir temos as funções que serão utilizadas para realizar conversões, soma e subtração, pegar parte do dado para o Código de Operação e Operando: 

```python
# Função que converte binário em decimal
def conversorBinHex(binario):
    h = ""
   
    cont = 0
    for i in range(len(binario)):
      cont += 1
    if(cont == 4): # Parte que converte o Endereço de 4 bits de binário para Hexadecimal
        if(binario == "0000"):
            h += "0"
        elif(binario == "0001"):
            h += "1"
        elif(binario == "0010"):
            h += "2"
        elif(binario == "0011"):
            h += "3"
        elif(binario == "0100"):
            h += "4" 
        elif(binario == "0101"):
            h += "5"
        elif(binario == "0110"):
            h += "6"
        elif(binario == "0111"):
            h += "7"
        elif(binario == "1000"):
            h += "8"
        elif(binario == "1001"):
            h += "9"
        elif(binario == "1010"):
            h += "A"
        elif(binario == "1011"):
            h += "B"
        elif(binario == "1100"):
            h += "C"
        elif(binario == "1101"):
            h += "D"
        elif(binario == "1110"):
            h += "E"
        elif(binario == "1111"):
            h += "F"
    elif(cont == 12): # Parte que converte o Dado de 12 bits de binário para Hexadecimal
        
        for i in range(0,len(binario),4):
            if(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0000"):
                h += "0"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0001"):
                h += "1"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0010"):
                h += "2"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0011"):
                h += "3"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0100"):
                h += "4"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0101"):
                h += "5"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0110"):
                h += "6"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "0111"):
                h += "7"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1000"):
                h += "8"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1001"):
                h += "9"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1010"):
                h += "A"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1011"):
                h += "B"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1100"):
                h += "C"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1101"):
                h += "D"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1110"):
                h += "E"
            elif(binario[i] + binario[i+1] + binario[i+2] + binario[i+3] == "1111"):
                h += "F"

    return h
# Converte decimal em binário
def converterDecBin(dec):
    b = ""
    if(dec == 0):
        b += "0000"
    elif(dec == 1):
        b += "0001"
    elif(dec == 2):
        b += "0010"
    elif(dec == 3):
        b += "0011"
    elif(dec == 4):
        b += "0100"
    elif(dec == 5):
        b += "0101"
    elif(dec == 6):
        b += "0110"
    elif(dec == 7):
        b += "0111"
    elif(dec == 8):
        b += "1000"
    elif(dec == 9):
        b += "1001"
    elif(dec == 10):
        b += "1010"
    elif(dec == 11):
        b += "1011"
    elif(dec == 12):
        b += "1100"
    elif(dec == 13):
        b += "1101"
    elif(dec == 14):
        b += "1110"
    elif(dec == 15):
        b += "1111"
    return b

# Converter binario em dec

def binDec(bd):
    d = 0
    for i in range(len(bd)):
        pot = len(bd)
    
    for i in range(len(bd)):
        pot += -1
        d = d + (int(bd[i])) * 2 ** pot
    return d

# Função para quebrar as partes do Dado em Operação
def quebraDadoOperacao(d):
    d1 = ""  # d1 --------------> 4 bits mais significativos para a Operação
    for i in range(len(d)):
        if(i >= 0 and i < 4):
            d1 += d[i]
    return(d1)

# Função para receber a parte da Instrução

def quebraDadoOperando(di):
    d3 = ""
    for i in range(len(di)):

        if(i >= 8 and i < 12):
            d3 += di[i]
    return(d3)
# Função para realizar a soma do binário

def somarBin(r,s):
    global resto
    if resto == "1":
        if (r == "1" and s == "0") or (r == "0" and s == "1"):
            resto = "1"
            return "0"
        elif r == "0" and s == "0":
            resto = "0"
            return "1"
        elif r == "1" and s == "1":
            resto = "1"
            return "1"
    else:
        if (r == "1" and s == "0") or (r == "0" and s == "1"):
            return "1"
        elif r == "0" and s == "0":
            return "0"
        elif r == "1" and s == "1":
            resto = "1"
            return "0"
# Função para subtrair o binário        
def subBin(r, s):
    print(r + " - " + s)

    c1 = ""
    for i in range(len(s)):
        if s[i] == "0":
            c1 += "1"
        else:
            c1 += "0"
    print("\nBinário: "+s+"\nComplemento de 1: "+c1)

    um = "1"
    while len(c1) > len(um):
        um = "0" + um

    global resto
    resto = "0"
    c2 = ""
    for i in range(len(c1)-1, -1, -1):
        c2 = somarBin(c1[i], um[i]) + c2
    print("Complemento de 2: "+c1+" + 1 = "+c2)

    sub = ""
    for i in range(len(c2)-1, -1, -1):
        sub = somarBin(r[i], c2[i]) + sub
    print("\nSoma final: \n"+r + " + " + c2+ " = "+str(sub))
  
    return sub
resto = ""

```

### A seguir temos a função que formata o Endereço:

```python
# Função para quebrar as partes do endereço

def formatoEndereco(endereco):
   
    global Ltag  # String para receber o endereço em binário correspondente ao Campo TAG
    global Lconjunto  # String para receber o endereço em binário correspondente ao Campo conjunto
    global Lbyte  # String para receber o endereço em binário correspondente ao Campo byte

    Ltag = ""
    Lconjunto = ""
    Lbyte = ""
    for i in range(len(endereco)):
        if(i>=0 and i<larguraCampoTag): # Separa o bit do campo Tag
            Ltag = Ltag + endereco[i]
        elif(i>=larguraCampoTag and i<(larguraCampoTag+ larguraCampoConjunto)): # Separa o bit do campo conjunto
            Lconjunto = Lconjunto + endereco[i]
        elif(i>=(larguraCampoTag+ larguraCampoConjunto)): # Separa o bit do campo byte
            Lbyte = Lbyte + endereco[i]
                 
    print(" ")
    #Imprime o bit de cada campo     
    print("\nFormato de endereço da cache: \n| Tag   | |Conjunto  | |Byte  | ")
    print("  ", Ltag,  "       " ,   Lconjunto,     "         " , Lbyte)
    print("")
    bloco =  (Ltag) + (Lconjunto)
    print("O bloco que será procurado é: ", bloco)
              
```

### E em seguida, temos  a Função responsável por realizar as Operações do Processador, de acordo com as instruções dadas:

```python

# FUNÇÃO PARA REALIZAR O QUE O PROCESSADOR FAZ
def operarProcessador():
    global RDM
    global REM
    global RI
    global PC
    global R0

    RDM = ""
    REM = ""
    RI = ""
    PC = ""
    R0 = ""
    print("------------ INICIANDO PROGRAMA ------------")
    PC = "0000"
    Ciclo = 0
    while True:
        Ciclo += 1
        print("Ciclo: ", Ciclo)
        REM = PC
        print("REM: ", REM)
        PC = converterDecBin(binDec(PC)  + 1)
        RDM = lerMemoria(REM)
        print("RDM: ", RDM)
        RI = RDM
        print("RI: ", RI)
    
        print("Próxima Instrução ------> ",PC)
        print("")
        
        if(conversorBinHex(quebraDadoOperacao(RI)) == "1"): # C.Op = 1 <----- LOAD - CARREGANDO
            print("LDA Op")
            REM = quebraDadoOperando(RI)
            print("REM: ", REM)
            RDM = lerMemoria(REM)
            print("RDM: ", RDM)
            R0 = RDM
            print("BIN        | ", "| HEX |")
            print(R0, "  ", conversorBinHex(R0))

        elif(conversorBinHex(quebraDadoOperacao(RI)) == "4"): # C.Op = 4 <----- SUB- SUBTRAÇÃO
            print("SUB Op")
            REM = quebraDadoOperando(RI)
            print("REM: ", REM)
            RDM = lerMemoria(REM)
            print("RDM: ", RDM)
            if binDec(RDM) > binDec(R0):
                flag = 1 # 1 <------- Resultado negativo
            else:
                flag = 0 # 0 <------- Resultado Positivo
            R0 = subBin(R0, RDM) # R0 recebe a subtração de R0 pelo dado do endereço do operando
            print("R0  <---  | "+R0+" | "+conversorBinHex(R0)+" |")
        elif(conversorBinHex(quebraDadoOperacao(RI)) == "6"): # C.Op = 6 <-------- JP - R0 > 0
            print("JP Op") 
            if(flag == 0): # R0 > 0 <----- Positivo
                PC = quebraDadoOperando(RI)
                print("Jump realizado\nPC <---  End | "+PC+" | "+conversorBinHex(PC)+" |")
            else:
                print("Jump não realizado")
        elif(conversorBinHex(quebraDadoOperacao(RI)) == "A"): # C.Op = A <-------- PRT - Printa o valor armazenado em (Op)
                print("PRT Op")
                REM = quebraDadoOperando(RI)
                print("REM ", REM)
                RDM = lerMemoria(REM)
                print("RDM ", RDM)
                print("End "+conversorBinHex(REM)+"    | "+RDM+" | "+conversorBinHex(RDM)+" |")

        elif(conversorBinHex(quebraDadoOperacao(RI)) == "8"): # C.Op = 8 <------ JMP - PC <-- Op
            print("JMP Op")
            PC = quebraDadoOperando(RI) # PC recebe endereço do operando obtido pela função para realizar o salto
            print("Jump realizado\nPC <---  End | "+PC+" | "+conversorBinHex(PC)+" |")

        elif(conversorBinHex(quebraDadoOperacao(RI)) == "2"): # C.Op = 2 <----------- STR - STORE <-------- (Op) <-- R0
            print("STR Op")
            REM = quebraDadoOperando(RI)
            print("REM: ", REM)
            RDM = R0
            print("RDM: ", RDM)
            escreverMemoria(REM , RDM) 
            print("End "+ conversorBinHex(REM)+"   |"+ R0+" | "+ conversorBinHex(R0)+" |")
        elif(conversorBinHex(quebraDadoOperacao(RI)) == "5"):
            print("JZ Op")
            if(R0 == "000000000000"):
                PC = quebraDadoOperando(RI) # PC recebe endereço do operando obtido pela função para realizar o salto
                print("Jump realizado\nPC <---  End | "+PC+" | "+conversorBinHex(PC)+" |")
            else:
                print("Jump não realizado")
        elif(conversorBinHex(quebraDadoOperacao(RI)) == "0"): # C.Op = 0
            print("HLT - Encerra a execução do programa")
            break
    print("------------ PROGRAMA ENCERRADO ------------")
        

```
